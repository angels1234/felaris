```python
 
from openpyxl import *
from tkinter import *


fel = load_workbook('alpha.xlsx') 


sheet = fel.active 


def excel(): 

	sheet.column_dimensions['A'].width = 35
	sheet.column_dimensions['B'].width = 15
	sheet.column_dimensions['C'].width = 15
	sheet.column_dimensions['D'].width = 30
	sheet.column_dimensions['E'].width = 30
	sheet.column_dimensions['F'].width = 50
	sheet.column_dimensions['G'].width = 55


	sheet.cell(row=1, column=1).value = "Name"
	sheet.cell(row=1, column=2).value = "Course"
	sheet.cell(row=1, column=3).value = "Year"
	sheet.cell(row=1, column=4).value = "Reference Number"
	sheet.cell(row=1, column=5).value = "Contact Number"
	sheet.cell(row=1, column=6).value = "Email id"
	sheet.cell(row=1, column=7).value = "Index Number"



def focus1(event): 
	course_field.focus_set() 



def focus2(event): 
	year_field.focus_set() 



def focus3(event): 
	reference_no_field.focus_set() 



def focus4(event): 
	contact_no_field.focus_set() 


def focus5(event):  
	email_id_field.focus_set() 


def focus6(event): 
	index_no_field.focus_set() 


def clear(): 

	name_field.delete(0, END) 
	course_field.delete(0, END) 
	year_field.delete(0, END) 
	reference_no_field.delete(0, END) 
	contact_no_field.delete(0, END) 
	email_id_field.delete(0, END) 
	index_no_field.delete(0, END) 


def insert(): 
	
	
	if (name_field.get() == "" and
		course_field.get() == "" and
		year_field.get() == "" and
		reference_no_field.get() == "" and
		contact_no_field.get() == "" and
		email_id_field.get() == "" and
		index_no_field.get() == ""): 

		print("Empty Input") 

	else: 

 
		current_row = sheet.max_row 
		current_column = sheet.max_column 

        
		sheet.cell(row=current_row + 1, column=1).value = name_field.get() 
		sheet.cell(row=current_row + 1, column=2).value = course_field.get() 
		sheet.cell(row=current_row + 1, column=3).value = year_field.get() 
		sheet.cell(row=current_row + 1, column=4).value = reference_no_field.get() 
		sheet.cell(row=current_row + 1, column=5).value = contact_no_field.get() 
		sheet.cell(row=current_row + 1, column=6).value = email_id_field.get() 
		sheet.cell(row=current_row + 1, column=7).value = index_no_field.get() 


		fel.save('alpha.xlsx') 


		name_field.focus_set() 


		clear() 

if __name__ == "__main__": 
	

	root = Tk() 


	root.configure(background='yellow') 


	root.title("Registration Form for Physics") 

 
	root.geometry("550x350") 

	excel() 


	heading = Label(root, text="Physics Forms", bg="yellow") 


	name = Label(root, text="Name", bg="yellow") 

 
	course = Label(root, text="Course", bg="yellow") 


	year = Label(root, text="Year", bg="yellow") 


	reference_no = Label(root, text="Ref No.", bg="yellow") 


	contact_no = Label(root, text="Contact No.", bg="yellow") 

 
	email_id = Label(root, text="Email ", bg="yellow") 


	index_no= Label(root, text="Index No.", bg="yellow") 
 
	heading.grid(row=0, column=1) 
	name.grid(row=1, column=0) 
	course.grid(row=2, column=0) 
	year.grid(row=3, column=0) 
	reference_no.grid(row=4, column=0) 
	contact_no.grid(row=5, column=0) 
	email_id.grid(row=6, column=0) 
	index_no.grid(row=7, column=0) 

 
	name_field = Entry(root) 
	course_field = Entry(root) 
	year_field = Entry(root) 
	reference_no_field = Entry(root) 
	contact_no_field = Entry(root) 
	email_id_field = Entry(root) 
	index_no_field = Entry(root) 


	name_field.bind("<Return>", focus1) 


	course_field.bind("<Return>", focus2) 


	year_field.bind("<Return>", focus3) 


	reference_no_field.bind("<Return>", focus4) 

 
	contact_no_field.bind("<Return>", focus5) 


	email_id_field.bind("<Return>", focus6) 

	name_field.grid(row=1, column=1, ipadx="130") 
	course_field.grid(row=2, column=1, ipadx="130") 
	year_field.grid(row=3, column=1, ipadx="130") 
	reference_no_field.grid(row=4, column=1, ipadx="130") 
	contact_no_field.grid(row=5, column=1, ipadx="130") 
	email_id_field.grid(row=6, column=1, ipadx="130") 
	index_no_field.grid(row=7, column=1, ipadx="130") 


	excel() 

	
	submit = Button(root, text="Send", fg="Gold", 
							bg="Black", command=insert) 
	submit.grid(row=8, column=1) 

	
	root.mainloop() 
```
